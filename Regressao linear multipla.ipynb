# Treinamento de Regressão Linear Múltipla




## Problema



*   Estimativa de rendimento em shopping a partir de idade e e score de clientes



## Carregamento de dados

```
import pandas as pd
```

```
dados = pd.read_csv("https://raw.githubusercontent.com/ect-info/ml/master/dados/Mall_Customers.csv")
dados.head()
```

```
dados.describe()
```

```
#Variável independente
X = dados.iloc[: , [2,3]].values
Y = dados.iloc[: , [4]].values
```

```
X_train = X[0:50]
X_test = X[50:] 
```

```
Y_train = Y[0:50]
Y_test = Y[50:]
```

### Análise de correlação

```
import matplotlib.pyplot as plt
import seaborn as sns
sns.pairplot(dados)
plt.tight_layout()
```

```
dados.corr()
```

```
sns.heatmap(dados.corr())
```

## TREINAMENTO / AJUSTE DO MODELO

### Ajustes

```
from sklearn import linear_model
from sklearn.metrics import mean_squared_error, r2_score


regr = linear_model.LinearRegression()
```

```
regr.fit(X_train,Y_train)

Y_pred = regr.predict(X_test)

print("Mean squared error: %.2f"
      % mean_squared_error(Y_test, Y_pred))
```

```
print('Variance score: %.2f' % r2_score(Y_test, Y_pred))
```

```
print('Coefficients: \n', regr.coef_)
```

 **Os dados utilizados no treinamento e teste foram idades dos clientes, rendimento anual e seus scores**

 **O erro quadrático médio não foi alto, logo o y_pred teve uma discrepância pequena quando associado com o y_test. Dessa vez também consegui calcular a variância e os coeficientes de regressão.**
